```python
import cv2  # OpenCV library
import os   # Directory handling

# Load Haar Cascade for face detection
haar_file = "haarcascade_frontalface_default.xml"
datasets = "dataset"
sub_data = "steve"

# Create dataset folder if it doesn't exist
path = os.path.join(datasets, sub_data)
if not os.path.isdir(path):
    os.makedirs(path)

# Set face image size
(width, height) = (130, 100)

# Load Haar cascade
face_cascade = cv2.CascadeClassifier(haar_file)
if face_cascade.empty():
    print("Error: Haar cascade file not loaded properly!")
    exit()

# Initialize webcam
webcam = cv2.VideoCapture(0)

# Check if the webcam is opened
if not webcam.isOpened():
    print("Error: Could not access the webcam")
    exit()

count = 1

while count < 31:
    print(f"Capturing image {count}")

    ret, im = webcam.read()  # Read frame from webcam

    if not ret or im is None:
        print("Error: Failed to capture image")
        continue  # Skip this loop iteration if the frame is invalid

    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    faces = face_cascade.detectMultiScale(gray, 1.3, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(im, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw rectangle
        face = gray[y:y+h, x:x+w]  # Crop the detected face
        face_resize = cv2.resize(face, (width, height))  # Resize face

        cv2.imwrite(f"{path}/{count}.png", face_resize)  # Save image
        count += 1

    cv2.imshow("Face Detection", im)

    # Press 'ESC' to exit
    key = cv2.waitKey(10)
    if key == 27:
        break

# Release the webcam and close OpenCV windows **outside the loop**
webcam.release()
cv2.destroyAllWindows()
```
